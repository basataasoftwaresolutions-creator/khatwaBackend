```
conversation = []

def ask_ai(prompt):
    response = model.generate_content(prompt)
    return response.text



# -------- Stage 1: جمع البيانات --------

def collect_user_data():
    data = {}

    data["budget"] = input("ما هي الميزانية المتاحة؟ ")
    data["location"] = input("في أي محافظة أو مدينة؟ ")
    data["experience"] = input("هل لديك خبرة سابقة؟ وفي أي مجال؟ ")
    data["time"] = input("كم ساعة تستطيع العمل يوميًا؟ ")
    data["goal"] = input("ما هدفك من المشروع؟ ")

    return data


# -------- Stage 2: التحليل --------

def analyze_user(data):
    prompt = f"""
أنت محلل أعمال محترف.

بناءً على هذه البيانات:
الميزانية: {data['budget']}
المكان: {data['location']}
الخبرة: {data['experience']}
الوقت المتاح: {data['time']}
الهدف: {data['goal']}

قم بتحليل الحالة واقترح أنواع مشاريع مناسبة.
"""
    return ask_ai(prompt)


# -------- Stage 3: Business Plan --------

def generate_business_plan(analysis):
    prompt = f"""
أنت خبير تخطيط مشاريع.

بناءً على هذا التحليل:
{analysis}

أنشئ خطة مشروع كاملة تشمل:
- فكرة المشروع
- التكاليف المتوقعة
- الأرباح المحتملة
- المخاطر
- خطة 3 شهور خطوة بخطوة
"""
    return ask_ai(prompt)


# -------- Stage 4: التقييم --------

def evaluate_project(plan):
    prompt = f"""
أنت مستشار استثماري.

قيّم هذه الخطة من حيث:
الربحية
المخاطرة
سهولة التنفيذ

وأعطِ تقييم نهائي مع السبب:
{plan}
"""
    return ask_ai(prompt)


# -------- Stage 5: التسويق --------

def generate_marketing(plan):
    prompt = f"""
أنت خبير تسويق رقمي.

أنشئ خطة تسويق للمشروع تشمل:
اسم براند
جمهور مستهدف
أفكار إعلانات
محتوى شهر كامل لتك توك وإنستجرام

بناءً على هذه الخطة:
{plan}
"""
    return ask_ai(prompt)


# -------- تشغيل النظام --------

def run_ai_consultant():
    user_data = collect_user_data()

    analysis = analyze_user(user_data)
    print("\n📊 التحليل:\n", analysis)

    plan = generate_business_plan(analysis)
    print("\n📈 خطة المشروع:\n", plan)

    evaluation = evaluate_project(plan)
    print("\n⭐ التقييم:\n", evaluation)

    marketing = generate_marketing(plan)
    print("\n📢 خطة التسويق:\n", marketing)


run_ai_consultant()
```
